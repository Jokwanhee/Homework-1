## 1. 다음 문제를 읽고 알맞은 코드를 제시하시오.

- 1) 파일 s.txt 내의 다음 내용과 함께 아래 문제의 코드를 제시하시오.
```
pig ham
cat dog
ham bird
dog pig
```

````python
f = open("s.txt",'w')
f.write("""pig ham
cat dog
ham bird
dog pig
""")
f.close()

f = open("s.txt", 'r')
print(f.read())
f.close()
````

 f로 s.txt파일을 w모드로 작성하였다. 여기서 "(따옴표) 3개를 사용하여 엔터를 친 후 줄바꿈을 사용해서 각 라인에 기입되게 하였다. 

- 2) 파일 s.txt을 읽어서 각 라인에 있는 첫 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s1.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s1.txt에 있는 파일 내용은 다음과 같다.
```
cat dog
dog pig
ham bird
pig ham
```

````python
f = open("s.txt",'r')
my_list = f.readlines()
my_list.sort()
f.close()

f = open("s1.txt",'w')
for i in range(len(my_list)):
    f.write(my_list[i])
f.close()

f = open("s1.txt",'r')
print(f.read())
f.close()
````

 우선 s.txt파일의 내용을 읽어온 후 my_list 변수 = f.readlines()를 저장한다. 그러면 my_list는 리스트 형태로 각 라인을 순서대로 저장한다. 이것을 첫번째 단어로 정렬하기 위해서는 sort 메소드를 이용해주면 된다. 저장된 my_list를 w모드로 s1.txt파일을 만든 후 리스트 크기만큼 각 리스트 항목을 작성해주면 첫 번째 단어로 정렬된다.

- 3) 파일 s.txt을 읽어서 각 라인에 있는 두 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s2.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s2.txt에 있는 파일 내용은 다음과 같다.
```
ham bird
cat dog
pig ham
dog pig
```

````python
f = open("s.txt",'r')
my_list = []
for line in f:
    extra = line.split()
    my_list.append(extra)
    my_list.sort(key = lambda x:x[1])
f.close()

f = open("s2.txt",'w')
new_list = []
for i in my_list:
    i = ' '.join(i)
    new_list += i + "\n"
f.writelines(new_list)
f.close()

f = open("s2.txt",'r')
print(f.read())
f.close()
````

 우선 s.txt파일을 읽어온다. list라는 빈 리스트를 만든 후 불러 온 파일의 각 라인을 알 수 있게 하기 위해서 반복문을 사용한다. 각 라인을 띄어쓰기가 있어서 단어로 분리되게 만들어 extra 변수에 저장했다. 그리고 그 extra 변수를 아까 만든 빈 리스트 list에 append한다. 그리고 그 list에서 sort 메소드를 활용하여 key = lambda x:x[1] 로 두 번째 단어의 순서대로 정렬할 수 있게 각 리스트를 항목을 정렬한다.   
 그 다음으로 s2.txt파일을 w모드로 작성해야 한다. new_list라는 빈 리스트를 만들고 위에서 정렬한 lsit를 불러와 for 반복문으로 한 라인을 각각 불러온다. 그리고 2차원 배열형태로 list를 정렬되어 있기 때문에 join메소드를 이용해서 각 리스트의 2차원 배열을 문자열로 만들고, 그 문자열을 new_list의 더한다. 그리고 끝에 개행문자를 만들어 줄바꿈을 해줘야 한다. 그 것을 writelines로 한번에 정리할 수 있다. 

- 4) 파일 s.txt을 읽어서 각 라인들에 있는 모든 단어들을 순차적으로 다시 나열하되 각 라인에 세 개의 단어들이 오도록 하여 s3.txt에 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s3.txt에 있는 파일 내용은 다음과 같다.
```
pig ham cat
dog ham bird
dog pig
```

````python
f = open("s.txt",'r')
me_list = f.readlines()
f.close()

f = open("s3.txt",'w')
str_list = ''.join(me_list)
sort_list = str_list.split()
count = -1
end = 1

while True:
    my_list = []
    
    for i in range(3):
        count += 1
        if len(sort_list) == count:
            end = -1
            break
        else:
            my_list += sort_list[count] + ' '
    my_list += '\n'
    
    f.writelines(my_list)
        
    if end == -1:
        break
              
f.close()

f = open("s3.txt",'r')
print(f.read())
f.close()
````

 우선 s.txt를 r모드로 불러온다. list 리스트로 f.readlines()를 저장한다.   
 그 다음으로 s3.txt를 w모드로 불러와 작성할 수 있어야 한다. 우선 위에서 list를 join메소드를 이용해서 문자열로 만든다. 그리고 이것을 split메소드를 활용해서 각 단어로 정의해 리스트에 저장한다. while 무한 반복문을 사용했다. 우선 빈 리스트 my_list를 만들고 3단어씩 한줄에 보여주게 하기위해서 for 반복문을 3번 돌렸다. 여기서 count를 통해서 리스트 안에 단어의 개수와 count가 동일 시 end 변수에 -1를 저장하여 for문을 나온다. 그리고 while 조건문에서 end 가 -1이면 while문을 종료한다. 하지만 리스트 안 단어의 개수와 count가 같지 않으면 아까 만든 my_list에 각 단어들이 정렬들의 인덱스를 count로 사용하고, ' '띄어쓰기도 마지막에 넣어 저장해준다. 그리고 my_list가 모두 정의되면 끝에 개행문자를 더해준다. 그리고 my_list를 writelines를 통해 작성을 완료한다.

## 3. range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

- 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
>range(stop)   
range(start, stop[, step])

- 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.

- 3) frange 사용 예
    - 3-1) frange(0.5)
        - [0.0, 0.1, 0.2, 0.3, 0.4]
    - 3-2) frange(1.0, 2.0)
        - [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
    - 3-3) frange(2.2, 4.0, 0.5)
        - [2.2, 2.7, 3.2, 3.7]

````python
def frange(start=0.0, *stop, step=0.1):
    mylist = []
    if len(stop) == 0:
        stop = start
        start = 0.0
        while(start < stop):
            mylist.append(start)
            start = round(start + step, 10)
    elif len(stop) == 1:
        while(start < stop[0]):
            mylist.append(start)
            start = round(start + step, 10)
    elif len(stop) == 2:
        step = stop[1]
        while(start < stop[0]):
            mylist.append(start)
            start = round(start + step, 10)
    
    return mylist


print(frange(0.5))
print(frange(1.0,2.0))
print(frange(2.2,4.0,0.5))
````

frange함수를 만들어 start, stop, step의 인자를 받는 함수이다. 우선 frange에서 받아올 수 있는 경우가 3가지이다. 한 개의 인자만 받을 때, 두 개의 인자를 받을 때, 세 개의 인자를 받을 때로 나뉜다.   
일단 frange 함수의 인자들은 start와 step은 디폴트 매개변수로 0.0과 0.1을 가진다. 그리고 stop 가변인수로 저장되어 튜플을 받아온다.   
우선 첫 번째로 한 개의 인자만 받을 때이다. 조건은 stop이 받아온 개수가 0개일 때다. 이 조건이면 stop은 start에서 받아온 숫자를 저장하고, start는 0.0으로 초기화한다. 그리고 start가 stop보다 작을 때 mylist라는 빈 리스트에 start값을 넣는다. 여기서 다음 start는 step에 영향을 받는데, 여기서는 round메소드를 활용하여 실수값의 오차를 벗어나는 변수를 없애기 위해서 사용해준다. 기본 step은 0.1이기 때문에 한 개의 인자를 받을 경우 그 값이 stop이 되고 0.0부터 0.1씩 증가하는 값을 얻게된다.   
다음으로 두 개의 인자를 받는 경우이다. 그 경우 stop의 개수는 1개이고, start가 stop 튜플의 첫 번째 인덱스보다 작을 때, mylist에 start값을 저장한다. 그리고 이 경우 또한, step 값이 기본 0.1이기 때문에 0.1씩 증가한다. 하지만 start와 stop의 값은 정할 수 있다.   
마지막으로 세 개의 인자를 받는 경우이다. 그 경우에는 stop의 개수가 2개이며 step 의 값은 stop 튜플의 1번째 인덱스의 값이다. 그리고 start가 stop 튜플 인덱스 0 번째의 값보다 작을 때, start의 값을 mylist에 저장한다. 그 다음 start는 step값을 정한 값 만큼 더해진다.

## 4. 가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오

- sum() 사용 예


>sum()   
0   

>sum(1,2)   
3   

>sum(1,2,3,4,5)   
15   

>sum(1, 5, 7, 2, -10)   
5   

````python
def sum(*arg):
    arg_sum = 0
    for i in arg:
        arg_sum += i
    return arg_sum

print(sum())
print(sum(1,2))
print(sum(1,2,3,4,5))
print(sum(1,5,7,2,-10))
````

함수를 만들어 가변인자를 사용하였다. 튜플형태를 받아올 수 있는 *arg를 만들었다. 우선 arg_sum = 0 변수를 만들고, for 반복문을 사용해 arg에서 받아온 각각 숫자들을 arg_sum 에 계속 더해나간다. 그리고 return arg_sum을 하면 함수밖에서 함수를 호출할 때 사용한 인수들의 합을 알 수 있다.

## 5. 여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.



- 1) 다음에 제시되는 함수들을 모두 이용해야 한다.
    - split
    - map
    - join
- 2) myinitial() 함수 사용 예
    - myinitial("as soon as possible")
        - ASAP

````python
def myinitial(in_string):
    in_string = in_string.split()
        
    map_ord_string = list(map(lambda x: ord(x[0]) - 32, in_string))
    map_chr_string = list(map(lambda x: chr(x), map_ord_string))
    map_delete_string = list(map(lambda x: in_string[x][1:], range(len(in_string))))
    map_string = list(map(lambda x: map_chr_string[x] + map_delete_string[x], range(len(in_string))))

    map_string = " ".join(map_string)
    return map_string
    
print(myinitial("as soon as possible"))
print(myinitial("my name is jokwanhee"))
print(myinitial("h h h h"))
````

우선 upper()라는 메소드가 있었지만, 그것을 활용하는 것이 아닌 split, map, join을 사용했다. 우선 input_string을 split()을 사용해서 띄어쓰기가 된 각 단어들을 리스트로 저장한다. 그리고 map_ord_string이라는 변수에 map을 사용해서 각 단어의 첫 글자의 아스키 숫자에서 32를 빼면 알파벳 소문자에서 대문자로 변환할 수 있다. 아스키 코드를 활용하였다. 그리고 map_chr_string에는 map을 사용해서 아까 저장된map_ord_string에서 아스키 숫자를 아스키 문자로 알파벳 대문자로 변환해준다. 그 다음으로 map_delete_string으로 map을 사용해서 각 단어의 숫자 만큼 각 단어의 첫 글자를 제외한 데이터만 저장한다. 그리고 map_string을 이용해서 전에 첫 글자를 제외한 데이터를 가지고 있는 map_delete_string에 아까 map_chr_string으로 대문자를 변환했던 값들을 각 단어의 첫글자로 더해준다. 마지막으로 map_string을 join을 사용해서 문자열로 만들어 준다. 여기서 " ". 와 같이 띄어쓰기를 해준다.

## 6. 음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

- [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

````python
def myfact(n):
    if n == 1:
        return 1
    return n * myfact(n-1)
    
myfact(2958)
````
