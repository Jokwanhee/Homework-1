## 2. 다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오.

- 다형성은 상속 관계 내의 다른 클래스들의 인스턴스들이 같은 멤버 함수 호출에 대해 각각 다르게 반응하도록 하는 기능이다.

```python
class KIA:
    def start(self):
        print("KIA")
    def end(self):
        return "KIA"
        
class K1(KIA):
    pass
    
class K3(KIA):
    def start(self):
        print("부릉")
    def end(self, a, b):
        return a * b
        
class K5(KIA):
    def start(self):
        print("부릉부릉")
    def end(self, a, b):
        return a + b
        
class K7(KIA):
    def start(self):
        print("부릉부릉부릉")
    def end(self, a, b):
        return a - b


for member in (K1(),K3(),K5(),K7()):
    member.start()
print('-'*40)    

a = K3()
b = K5()
c = K7()
d = K1()
print(a.end("부릉",3))
print(b.end("K","5"))
print(c.end(10,3))
print(d.end())
```

위 코드는 다형성이다. 다형성은 다른 클래스에 속한 같은 이름의 인스턴스들이 동일한 메소드 이름으로 호출할 경우 동적으로 선택되어 수행한다. 즉, 다양한 형태를 가질 수 있는 것이다. KIA라는 부모클래스를 만들고, 자식클래스로는 K1, K3, K5, K7가 있다. 각 클래스에 같은 이름의 함수를 만들고 인스턴스로 그 함수를 불렀을 때, 각 설정하였던 출력값이 나오는 것을 알 수 있다. K1같은 경우에는 pass를 넣어 부모클래스의 출력값이 나온다. 그리고 각 자식클래스에 새로운 함수 end를 만들어 매개변수를 받을 수 있는 형태로 작성하고 인스턴스를 만들어 그 인스턴스로 함수를 불러서 매개변수를 넣었을 때, 어떻게 나오는지 확인해보았다.  

## 3. 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

- 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.
    - count: Counter 인스턴스가 지니는 초기 정수 값
    - step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)
    >  c = Counter(10)   
    > d = Counter(10, 2)
    
- 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.

- 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.

    d.incr()   
    print(c)   
    [Count (step: 1)] 11

- 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.
    > c()   
    d()   
    print(c)   
    [Count (step: 1)] 12   
    print(d)   
    [Count (step: 2)] 14
    
- 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
    > c = c + 5   
    d = d - 5   
    print(c)   
    [Count (step: 1)] 17   
    print(d)   
    [Count (step: 2)] 9

- 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
    > print(c > 10)   
    True   
    print(d > 10)   
    False   
    print(c < 10)   
    False   
    print(d < 10)   
    True   
    print(c == 17)   
    True   
    print(d != 9)   
    False

```python
class Counter:
    def __init__(self,count,step=1):
        self.count = count
        self.step = step
    def __call__(self):
        self.incr()
    def __str__(self):
        return "[Count (step:{})] {}".format(self.step, self.count)
    def incr(self):
        self.count += self.step
    def __add__(self, other):
        self.count += other
        return self
    def __sub__(self, other):
        self.count -= other
        return self
    def __eq__(self, other):
        if type(other) == Counter:
            return self.count == other.count
        else:
            return self.count == other
    def __ne__(self, other):
        if type(other) == Counter:
            return self.count != other.count
        else:
            return self.count != other
    def __gt__(self, other):
        if type(other) == Counter:
            return self.count > other.count
        else:
            return self.count > other
    def __lt__(self, other):
        if type(other) == Counter:
            return self.count < other.count
        else:
            return self.count < other
    def __ge__(self, other):
        if type(other) == Counter:
            return self.count >= other.count
        else:
            return self.count >= other
    def __le__(self, other):
        if type(other) == Counter:
            return self.count <= other.count
        else:
            return self.count <= other
    
    
        
c = Counter(10)
d = Counter(10,2)

print(c)
print(d)
print()

c.incr()
d.incr()
print(c)
print(d)
print()

c()
d()
print(c)
print(d)
print()

c = c + 5
d = d - 5
print(c)
print(d)
print()

print(c > 10)
print(d > 10)
print(c < 10)
print(d < 10)
print(c == 17)
print(d != 10)
```

Counter라는 클래스를 만들고, 인스턴스를 생성하여 매개변수를 받기 위해 생성자 __init__을 만들어주었다. 매개변수는 두개인데, count와 step이다. step은 디폴트 매개변수로 1의 값을 저장하고 있다. 그리고 인스턴스 변수를 만들고 그 변수를 print하였을 때, 출력되기 위하여 __str__메소드를 활용하여 문자열을 출력할 수 있게 한다. 다음으로는 인스턴스 변수.incr()함수를 통하여 각 count에 step만큼 더하는 함수를 만든다. 다음으로는 인스턴스 변수를 클래스의 객체로 호출할 수 있게 만들어주는 __call__메소드를 활용하여 그 안에 self.incr()함수를 넣음으로써 인스턴스 변수()를 호출할 경우 count에 step만큼 더할 수 있게끔 한다. 다음으로는 덧셈과 뺄셈을 하기 위하여 __add__와 __sub__를 통하여 진행한다. 다음으로는 각 인스턴스 변수와 정수값을 비교하기 위해서 eq, ne, gt, lt, ge, lt로 ==,!=,>,<,>=,<=의 비교연산자를 할 수 있게 설정하여 True와 False를 반환할 수 있게끔 코딩하였다.

## 4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

- MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list): 
    def __init__(self, l): 
        for e in l: 
            self.append(e) 
        MySet.eliminate_duplicate(self)                                         
    
    def __str__(self):
        result = "MySet: {"
        for e in self: 
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}" 
        return result 
    
    @staticmethod 
    def eliminate_duplicate(l): 
        s = [] 
        for e in l: 
            if e not in s:
                s.append(e) 
        l[:] = [] 
        for e in s: 
            l.append(e) 
    
if __name__ == "__main__": 
    s = MySet([1, 2, 2, 3]) 
    print(s) 
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t) 
```

안녕하세요. 오늘은 리스트를 집합형태로 바꾸는 클래스를 만들어 보겠습니다.   
우선 사용할 키포인트는 정적메소드 사용, 내장 자료형 list를 서브클래싱하는 것이라고 할 수 있겠습니다.   
이제 클래스를 만듭니다. 클래스 이름은 MySet이다. MySet은 내장 자료형 list를 서브클래싱하여 가지게 됩니다.   
- 처음으로 __init__ 생성자입니다.   
self, l 를 매개변수로 가지게 된다. 여기서 l은 불러온 내장 자료형 list의 항목들이 리스트의 형태로 저장되게 됩니다.   
for e in l: 반복문을 사용하여 받아온 list 가 [1,2,3]일 때, l = [1,2,3] e = 1..2..3을 가지게 됩니다. 반복문에서는 self.append(e)를 하는데, 클래스 밖에서 클래스를 호출하기 위한 인스턴스를 만들 때, 그 저장된 값이 self로 들어오게 된다. 여기서 self 는 새로운 list를 만드는 역할을 한다.   
다음으로 MySet.eliminate_duplicate(self)로 함수를 호출한다. self를 인자로 보내준다. 함수에서는 self를 리스트->집합의 형식으로 전환시킵니다.      
- 다음으로는 생성자에서 호출 한 eliminate_duplicate()함수를 알아보겠습니다.   
함수 위 @staticmethod가 있는데 그것에 대해 잠깐 설명하고 넘어가겠습니다.   
지금까지 클래스의 메소드를 사용할 때, myset = MySet()와 같이 인스턴스를 만들어서 호출했다. 하지만 인스턴스를 하지 않고 클래스에서 바로 호출할 수 있는 정적 메소드를 사용하였다. 정적 메소드는 메소드 위에 @staticemethod와 같이 작성한다. 이때 정적 메소드는 self를 지정하지 않는다. 그리고 앞서 배운 데코레이터와 같은 의미이다. 데코레이터는 기존 함수가 그대로 동작하면서 추가 문장이 더 실행되는 래퍼 함수를 의미한다.   
다시 함수로 넘어와서 eliminate_duplicate(l)함수는 받아온 self 리스트에서 중복된 값을 제거해주는 함수이다. l은 self로 저장된 리스트를 매개변수로 갖는다.   
s = [], s라는 빈 리스트 생성한다.   
for e in l: 반복문을 사용하여 l 리스트에서 각 항목들을 차례로 뽑아 반복한다. 만약 리스트 s에 l 리스트의 항목을 차례로 삽입하는 경우에 그 항목이 s에 존재하지 않을 경우에만 True 반환하여 s에 l 리스트 항목을 삽입시킨다.   
그 이후에 l 리스트 전체를 빈 리스트로 전환한다.   
for e in s: 다시 반복문을 사용하여 새로 만든 s 리스트의 항목들을 차례로 뽑아 반복한다. 이 반복문은 빈 리스트를 만든 l 을 다시 e 항목을 차례로 삽입하는 반복문이다.   
- 다음으로는 __str__ 어떤 객체를 문자열로 변환하는데 사용하는 것에 대해 알아보겠습니다.   
result = "MySet: {" 를 사용해 result에 MySet: { 문자열 저장
for e in self: 반복문을 사용하여 생성자에서 설정 된 리스트->집합으로 변환시킨 self를 각 항목별로 차례로 뽑아 반복한다. result에 result를 추가로 더하고, 항목 e를 문자열로 바꾸어 더한 후  , 를 더한다.   
result = result[0:len(result)-2] + "}" 를 사용해 다시 result를 마지막 " ," 빈칸+, 를 제거하고 }를 더하여 정리해준다.   
return result 출력될 문자열을 return으로 표현해준다. 여기서 __str__ 은 return 시 nontype으로 설정해야 한다.
- 마지막으로 클래스 밖 클래스 호출을 위한 방법입니다.
if __name__ == "__main__": 를 사용하여 이 파일의 메인 함수를 호출하기 위한 조건을 세웁니다. 여기서 __name__ 는 이미 있는 변수이고, 오직 이 파일에서만 사용가능하게 하여 메인 함수만 호출, 다른 파일로 넘어갈 시 오류가 발생합니다.

```python
class MySet(list): 
    def __init__(self, l):
        for e in l: 
            self.append(e) 
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self: 
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}" 
        return result 
    
    def __or__(self, other): # 합집합 '|'
        print("===합집합===")
        or_list = [] # or_list라는 빈 리스트 생성
        for x in self: # self 리스트의 항목을 불러오는 반복문
            or_list.append(x) # self 리스트 항목을 하나하나 or_list에 삽입
        for y in other: # other 리스트의 항목을 불러오는 반복문
            if y not in or_list: # or_list에 other 항목 중 특정항목이 없더라면 True 반환
                or_list.append(y) # or_list에 other 항목 삽입
        or_list.sort() # or_list 정렬
    
        result = "MySet: {"
        for e in or_list: 
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"   
        return result 
        
    def __and__(self, other): # 교집합 '&'
        print("===교집합===")
        and_list = [] # and_list 빈 리스트 생성
        for x in self: # self 항목을 가져오는 반복문
            and_list.append(x) # 그 항목을 and_list에 삽입
        for y in range(len(and_list)): # and_list 크기만큼 반복
            for z in and_list: # and_list 항목을 불러오는 반복문
                if z not in other: # 그 항목이 other에 들어있지 않다면
                    and_list.remove(z) # and_list에서 그 항목 제거
        and_list.sort() # and_list 정렬
        
        result = "MySet: {"
        for e in and_list: 
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}" 
        return result
        
        
        
    def __sub__(self, other): # 차집합 '-'
        print("===차집합===")
        sub_list = [] # sub_list 빈 리스트 생성
        for x in self: # self 항목을 불러오는 반복문
            sub_list.append(x) # 그 항목을 sub_list에 삽입
        for y in range(len(sub_list)): # sub_list 크기만큼 반복
            for z in sub_list: # sub_list 항목을 불러오는 반복문
                if z in other: # 그 항목이 other에 있으면
                    sub_list.remove(z) # 그 항목 sub_list에서 제거
        sub_list.sort() # self 리스트 정렬
        
        result = "MySet: {"
        for e in sub_list: 
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}" 
        return result
                

    @staticmethod 
    def eliminate_duplicate(l): 
        s = [] 
        for e in l: 
            if e not in s:
                s.append(e) 
        l[:] = [] 
        for e in s: 
            l.append(e) 
    
if __name__ == "__main__": 
    s = MySet([1, 2, 2, 3]) 
    print(s) 
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t) 
    u = s | t
    print(u)
    u = s & t
    print(u)
    u = s - t
    print(u)
```

    >w4.getWordsFrequency()   
    [('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]

    >w4.getWordsFrequency(reverse=true)   
    [('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]

```python
class SEWOWW(SearchEngine,list):
    def __init__(self, args):
        super(SEWOWW, self).__init__(args)
    def getWordsFrequency(self,reverse=False):
        super_dict = super(SEWOWW, self).getWordsFrequency()
        if not reverse: # 내림차순
            sort_dict = sorted(super_dict.items(), key=lambda x:x[1], reverse = True)
        else: # 오름차순
            sort_dict = sorted(super_dict.items(), key=lambda x:x[1], reverse=False)
        for i in sort_dict:
            self.append(i)
        return print(sort_dict,end="")
        
w4 = SEWOWW('https://kin.naver.com/')
w4.getWordsFrequency()
```

자식 클래스를 만들었다. 여기서는 불러온 딕셔너리를 내림차순과 오름차순으로 정렬하는 클래스이다. 이를 수행하기 위해서 reverse라는 매개변수를 설정하고 디폴트 값은 False로 설정하여 매개변수를 아무것도 입력하지 않을 경우 내림차순정렬이 되게 하였다. 내장함수 sorted를 이용하여 key, reverse를 이용하였고, key 설정은 람다함수를 이용하여 튜플 0번째가 아닌 1번째로 단어의 출현빈도 수의 맞추어 내림차순을 결정하였다. 매개변수로 True값을 받으면 반대로 오름차순으로 정렬되게 된다.
